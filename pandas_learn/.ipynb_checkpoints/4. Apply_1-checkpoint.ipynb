```python
import pandas as pd
import numpy
```

```python
# read_csv 可以使用相对路径
df = pd.read_csv("student-mat.csv",sep=",")
df.head()
```

```python
df.info()
```

```python
# loc 标签索引
sg = df.loc[:,"school":"guardian"]
sg.head(5)
```

```python
# 对某一列中所有值进行处理，通过apply，lambda是缺省函数
sg["Mjob"].apply(lambda x:x.upper()).head(5)
```

```python
# apply相当于scala中的map lambda函数也挺好用的
sg.Fjob.apply(lambda x:x.upper()).head()
```

```python
# 查看最后5行
sg.tail()
```

```python
sg.head()
```

```python
sg.Mjob = sg.Mjob.apply(lambda x:x.upper())
captalizer = lambda x:x.upper()
sg.Fjob = sg.Fjob.apply(captalizer)
sg.tail()
```

```python
def majority(x):
    if x>17:
        return True
    else:
        return False
```

```python
# 新增一列，对age做分组操作
sg["legal_drinker"]=sg.age.apply(majority)
sg.head()
```

```python
# 这种函数写法，可以避免传入错误的x类型导致的运行时错误，因为运行时Python检查不出来
def times10(x):
    if type(x) is str:
        return x
    elif type(x) is numpy.int64:
        return 10*x
    else:
        return
```

```python
# 对所有整个数据框做操作，这种可以整体做归一化处理用的到
sg.applymap(times10).head(10)
```
