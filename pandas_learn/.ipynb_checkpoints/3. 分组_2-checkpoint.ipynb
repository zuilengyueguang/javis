```python
import pandas as pd
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
```

```python
users.columns
```

```python
users.head()
```

```python
# 按照职业进行分组，展示年龄的均值
m = users.groupby("occupation").age.mean()
# 对分组数据进行取整处理
m.apply(round)
```

```python
# 计算某一列中所有类型的数目
u = users.occupation.value_counts().head()
print u
print type(u)
```

```python
a = users.groupby("occupation").age.sum()
print type(a)
# series 计算，按照行索引进行计算。相同的行索引会相加
a+u
```

```python
# 计算女士在职业中的占比
# 首先按照职业分组，计算职业数量和女士数量，然后让女士人数除以职业总人数
# users['age_new'] 可以添加新的列 users.age_new 不能添加新的列
#users['age_new']=1
#users.head(5)
# 进行类别和数值的转换
def gender_to_num(x):
    if x=="M":
        return 1
    if x=="F":
        return 0
# 添加新的列，存放顺序数据
users["gender_n"] = users["gender"].apply(gender_to_num)
# 按照职业分组，并计算女士数量，求得百分比
a = users.groupby("occupation").gender_n.sum()/users.occupation.value_counts()*100
# 倒序排序
a.sort_values(ascending=False)
```

```python
# 按照职业分组，查看年龄的最大值 最小值
users.groupby("occupation").age.agg(["min","max"])
```

```python
b = users.groupby("occupation").age
# 查看单列所有属性
print b.describe()
# 分组查看所有列均值
users.groupby("occupation").mean()
```

```python
# 按照两列分组查看均值
users.groupby(["occupation","gender"]).age.mean()
```

```python
# 按照职业 年龄分组，并查看年龄的数量
users.groupby(["occupation","gender"]).age.count()
# 按照职业 年龄分组，查看年龄的数量，并给定列标签
# agg是添加标签函数，不添加标签时 和 直接.count用法相同
gender_ocup = users.groupby(["occupation","gender"]).agg({"gender":"count"})
# 按照职业分组，得到所有列的数量
occup_count = users.groupby(["occupation"]).agg("count")
# 职业和年龄分组后的数据除以 按照职业分组后的数据 得到百分比
occup_gender = gender_ocup.div(occup_count,level="occupation")*100
# 只查看 gender列
occup_gender.loc[:,"gender"]
```
