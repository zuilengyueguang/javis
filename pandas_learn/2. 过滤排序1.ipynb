```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url,sep='\t')
```

```python
chipo.info()
```

```python
chipo.item_price = [float(value[1:-1]) for value in chipo.item_price]
```

```python
chipo10 = chipo[chipo.item_price>10.00]
print len(chipo10)
chipo10.head(3)
```

```python
print len(chipo.item_name.unique())
print type(chipo.item_name.unique()),chipo.item_name.unique()[1]
```

```python
# 去重函数，将item_name重复的那一行去掉，默认只保留第一行，第二行算是重复，去掉
chipo.drop_duplicates(['item_name'])
```

```python
# 去重函数
chipo.drop_duplicates(['quantity'])
```

```python
# 将item_name quantity都重复的行去掉
chipo.drop_duplicates(['item_name','quantity'])
```

```python
chipo_filtered = chipo.drop_duplicates(['item_name','quantity'])
chipo_filtered[chipo_filtered.quantity==1]
```

```python
# 去重
chipo_filtered = chipo.drop_duplicates(['item_name','quantity'])
# 过滤出quantity==1的数据
chipo_one_prod = chipo_filtered[chipo_filtered.quantity==1]
# 选取对应的列
price_per_item = chipo_one_prod[['item_name','item_price']]
# 按照item_price 正序排序  ascending=False为倒序
price_per_item.sort_values(by="item_price",ascending=True)
```

```python
# item_name进行排序
print chipo.item_name.sort_values()
# 整个dataframe按照item_name进行排序
chipo.sort_values(by="item_name")
```

```python
# chipo 按照 item_price倒序排序，并查看第一行
chipo.sort_values(by="item_price",ascending=False).head(1)
```

```python
# 一个过滤条件
chipo[chipo.item_name=='Veggie Salad Bowl']
```
